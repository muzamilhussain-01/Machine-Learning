```python
df = pd.DataFrame(dataset.data, columns=dataset.feature_names)
df.head()
```

```python
# here x is our dataframe and y is our target classes
x = df
y = dataset.target
```

```python
x
```

```python
y
```

```python
from sklearn.preprocessing import StandardScaler

scaler = StandardScaler()
x_scaled = scaler.fit_transform(x)
x_scaled
```

```python
from sklearn.model_selection import train_test_split

x_train, x_test, y_train, y_test = train_test_split(x_scaled, y , test_size=0.2, random_state=30)
```

```python
from sklearn.linear_model import LogisticRegression

model = LogisticRegression()
model.fit(x_train, y_train)
```

```python
model.score(x_test, y_test)
```

```python
#Now jump into the most important thing that is PCA lets get started
from sklearn.decomposition import PCA

#here we also can pass the "n_components=2 whatever value you want"  but if we are passing "0.95" so we are saying that give us 95% of important info
#from the dataset.  it will say to PCA that retain 95% of usefull features
pca = PCA(0.95)

x_pca = pca.fit_transform(x)
```

```python
# see you have got the 29 columns. originally we had 64 columns. in actual what PCA does is he gave us most usefull features 
x_pca.shape
```

```python
#what actualy it does is that it calculated neew columns for you
x_pca
```

```python
pca.explained_variance_ratio_

#see it calculates each column and sees is the columns giving 95% important  features or info or not
```

```python
pca.n_components_
```

```python
x_train_pca, x_test_pca, y_train, y_test = train_test_split(x_pca, y , test_size=0.2, random_state=30)
```

```python
model = LogisticRegression(max_iter=1000)
model.fit(x_train_pca, y_train)
```

```python
model.score(x_test_pca, y_test)
```
